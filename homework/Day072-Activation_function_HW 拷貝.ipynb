# Rectified Linear Unit- Relu 

f(x)=max(0,x)


```python
import numpy as np
from numpy import *
import matplotlib.pylab as plt
%matplotlib inline


def ReLu(x):
    return x if x>0 else 0

def dReLu(x):
    return 1 if x>0 else 0
```

```python
X = np.linspace(-3, 3, 260, endpoint=True)
Y = [ReLu(x) for x in X]
dY = [dReLu(x) for x in X]

plt.plot(X, Y)
plt.plot(X, dY)
plt.grid()
```
