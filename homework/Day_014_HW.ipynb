### 請根據不同的 HOUSETYPE_MODE 對 AMT_CREDIT 繪製 Histogram

```python
# Import 需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # 另一個繪圖-樣式套件

%matplotlib inline
plt.style.use('ggplot')
import warnings
warnings.filterwarnings('ignore')

# 設定 data_path
dir_data = '/Users/rogeredc/Documents/GitHub/ML100-Days-講義/part1/data'
```

```python
f_app = os.path.join(dir_data, 'application_train.csv')
print('Path of read in data: %s' % (f_app))
app_train = pd.read_csv(f_app)
app_train.head()
```

```python
app_train['HOUSETYPE_MODE'].unique()
```

```python
"""
Your Code Here
"""
app_train['HOUSETYPE_MODE'].fillna('na', inplace=True)
unique_house_type = np.sort(app_train['HOUSETYPE_MODE'].unique())

nrows = len(unique_house_type)
ncols = nrows // 2

plt.figure(figsize=(10,30))
for i in range(len(unique_house_type)):
    plt.subplot(nrows, ncols, i+1)
    """
    Your Code Here
    """
    app_train.loc[(app_train['HOUSETYPE_MODE'] == unique_house_type[i]) & \
                              (app_train['TARGET'] == 0), 'AMT_CREDIT'].hist(label="target=0")
    app_train.loc[(app_train['HOUSETYPE_MODE']== unique_house_type[i]) & \
                              (app_train['TARGET'] == 1), 'AMT_CREDIT'].hist(label="target=1")
    plt.title(str(unique_house_type[i]))
    plt.legend()
plt.show()    
```
